# Exploratory Data Analysis (EDA) - Hotel Reservations Dataset

This notebook performs comprehensive exploratory data analysis on the Hotel Reservations dataset.

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

## 2. Load Dataset

```python
# Load the dataset
df = pd.read_csv("Hotel Reservations.csv")
print("Dataset loaded successfully!")
print(f"Shape: {df.shape}")
```

## 3. Basic Dataset Information

```python
# Display first few rows
print("\n= First 5 Rows =")
df.head()
```

```python
# Display last few rows
print("\n=== Last 5 Rows ===")
df.tail()
```

```python
# Dataset info
print("\n=== Dataset Information ===")
df.info()
```

```python
# Statistical summary
print("\n=== Statistical Summary ===")
df.describe()
```

```python
# Check for missing values
print("\n=== Missing Values ===")
missing_values = df.isnull().sum()
print(missing_values)
print(f"\nTotal missing values: {missing_values.sum()}")
```

```python
# Check for duplicate rows
print("\n=== Duplicate Rows ===")
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

```python
# Data types
print("\n=== Data Types ===")
print(df.dtypes)
```

## 4. Univariate Analysis - Numerical Features

```python
# Select numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
# Remove Booking_ID as it's just an identifier
if 'Booking_ID' in numerical_cols:
    numerical_cols.remove('Booking_ID')

print("Numerical columns:", numerical_cols)
```

```python
# Distribution plots for numerical features
fig, axes = plt.subplots(4, 4, figsize=(20, 16))
axes = axes.ravel()

for idx, col in enumerate(numerical_cols):
    if idx < len(axes):
        axes[idx].hist(df[col], bins=30, edgecolor='black', alpha=0.7, color='skyblue')
        axes[idx].set_title(f'Distribution of {col}', fontsize=12, fontweight='bold')
        axes[idx].set_xlabel(col)
        axes[idx].set_ylabel('Frequency')
        axes[idx].grid(True, alpha=0.3)

# Hide extra subplots
for idx in range(len(numerical_cols), len(axes)):
    axes[idx].axis('off')

plt.tight_layout()
plt.show()
```

```python
# Box plots to detect outliers
fig, axes = plt.subplots(4, 4, figsize=(20, 16))
axes = axes.ravel()

for idx, col in enumerate(numerical_cols):
    if idx < len(axes):
        axes[idx].boxplot(df[col].dropna(), vert=True, patch_artist=True,
                         boxprops=dict(facecolor='lightblue', alpha=0.7))
        axes[idx].set_title(f'Box Plot of {col}', fontsize=12, fontweight='bold')
        axes[idx].set_ylabel(col)
        axes[idx].grid(True, alpha=0.3)

# Hide extra subplots
for idx in range(len(numerical_cols), len(axes)):
    axes[idx].axis('off')

plt.tight_layout()
plt.show()
```

## 5. Univariate Analysis - Categorical Features

```python
# Select categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
# Remove Booking_ID
if 'Booking_ID' in categorical_cols:
    categorical_cols.remove('Booking_ID')

print("Categorical columns:", categorical_cols)
```

```python
# Value counts for categorical features
for col in categorical_cols:
    print(f"\n=== {col} ===")
    print(df[col].value_counts())
    print(f"\nPercentage distribution:")
    print(df[col].value_counts(normalize=True) * 100)
```

```python
# Bar plots for categorical features
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
axes = axes.ravel()

for idx, col in enumerate(categorical_cols):
    if idx < len(axes):
        value_counts = df[col].value_counts()
        axes[idx].bar(range(len(value_counts)), value_counts.values, 
                     color='coral', alpha=0.7, edgecolor='black')
        axes[idx].set_xticks(range(len(value_counts)))
        axes[idx].set_xticklabels(value_counts.index, rotation=45, ha='right')
        axes[idx].set_title(f'Distribution of {col}', fontsize=12, fontweight='bold')
        axes[idx].set_ylabel('Count')
        axes[idx].grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

```python
# Pie charts for categorical features
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.ravel()

for idx, col in enumerate(categorical_cols):
    if idx < len(axes):
        value_counts = df[col].value_counts()
        axes[idx].pie(value_counts.values, labels=value_counts.index, autopct='%1.1f%%',
                     startangle=90, colors=sns.color_palette('pastel'))
        axes[idx].set_title(f'Proportion of {col}', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

## 6. Target Variable Analysis (Booking Status)

```python
# Booking status distribution
print("=== Booking Status Distribution ===")
print(df['booking_status'].value_counts())
print("\nPercentage:")
print(df['booking_status'].value_counts(normalize=True) * 100)
```

```python
# Visualize booking status
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Count plot
booking_counts = df['booking_status'].value_counts()
axes[0].bar(booking_counts.index, booking_counts.values, color=['green', 'red'], alpha=0.7, edgecolor='black')
axes[0].set_title('Booking Status Distribution', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Count')
axes[0].grid(True, alpha=0.3, axis='y')

# Pie chart
axes[1].pie(booking_counts.values, labels=booking_counts.index, autopct='%1.1f%%',
           startangle=90, colors=['green', 'red'], explode=[0.05, 0.05])
axes[1].set_title('Booking Status Proportion', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

## 7. Bivariate Analysis - Numerical vs Target

```python
# Box plots: Numerical features vs Booking Status
fig, axes = plt.subplots(4, 4, figsize=(20, 16))
axes = axes.ravel()

for idx, col in enumerate(numerical_cols):
    if idx < len(axes):
        df.boxplot(column=col, by='booking_status', ax=axes[idx], patch_artist=True)
        axes[idx].set_title(f'{col} vs Booking Status', fontsize=11, fontweight='bold')
        axes[idx].set_xlabel('Booking Status')
        axes[idx].set_ylabel(col)
        plt.sca(axes[idx])
        plt.xticks(rotation=45)

# Hide extra subplots
for idx in range(len(numerical_cols), len(axes)):
    axes[idx].axis('off')

plt.suptitle('')  # Remove the default title
plt.tight_layout()
plt.show()
```

```python
# Average values by booking status
print("=== Average Values by Booking Status ===")
grouped_stats = df.groupby('booking_status')[numerical_cols].mean()
print(grouped_stats)
```

## 8. Bivariate Analysis - Categorical vs Target

```python
# Stacked bar charts for categorical features vs booking status
cat_cols_to_plot = [col for col in categorical_cols if col != 'booking_status']

fig, axes = plt.subplots(len(cat_cols_to_plot), 1, figsize=(14, 5*len(cat_cols_to_plot)))
if len(cat_cols_to_plot) == 1:
    axes = [axes]

for idx, col in enumerate(cat_cols_to_plot):
    crosstab = pd.crosstab(df[col], df['booking_status'], normalize='index') * 100
    crosstab.plot(kind='bar', stacked=True, ax=axes[idx], color=['green', 'red'], alpha=0.7)
    axes[idx].set_title(f'{col} vs Booking Status (Percentage)', fontsize=12, fontweight='bold')
    axes[idx].set_xlabel(col)
    axes[idx].set_ylabel('Percentage')
    axes[idx].legend(title='Booking Status')
    axes[idx].grid(True, alpha=0.3, axis='y')
    plt.sca(axes[idx])
    plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
# Count plots for categorical features vs booking status
fig, axes = plt.subplots(len(cat_cols_to_plot), 1, figsize=(14, 5*len(cat_cols_to_plot)))
if len(cat_cols_to_plot) == 1:
    axes = [axes]

for idx, col in enumerate(cat_cols_to_plot):
    crosstab = pd.crosstab(df[col], df['booking_status'])
    crosstab.plot(kind='bar', ax=axes[idx], color=['green', 'red'], alpha=0.7, edgecolor='black')
    axes[idx].set_title(f'{col} vs Booking Status (Count)', fontsize=12, fontweight='bold')
    axes[idx].set_xlabel(col)
    axes[idx].set_ylabel('Count')
    axes[idx].legend(title='Booking Status')
    axes[idx].grid(True, alpha=0.3, axis='y')
    plt.sca(axes[idx])
    plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

## 9. Correlation Analysis

```python
# Correlation matrix
correlation_matrix = df[numerical_cols].corr()
print("=== Correlation Matrix ===")
print(correlation_matrix)
```

```python
# Heatmap of correlations
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Heatmap of Numerical Features', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```

```python
# Find highly correlated features (correlation > 0.7)
print("\n=== Highly Correlated Features (|correlation| > 0.7) ===")
high_corr_pairs = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        if abs(correlation_matrix.iloc[i, j]) > 0.7:
            high_corr_pairs.append((
                correlation_matrix.columns[i],
                correlation_matrix.columns[j],
                correlation_matrix.iloc[i, j]
            ))

if high_corr_pairs:
    for pair in high_corr_pairs:
        print(f"{pair[0]} <-> {pair[1]}: {pair[2]:.3f}")
else:
    print("No highly correlated features found.")
```

## 10. Pairplot for Key Features

```python
# Select key features for pairplot
key_features = ['lead_time', 'avg_price_per_room', 'no_of_special_requests', 
                'no_of_adults', 'no_of_weekend_nights', 'no_of_week_nights', 'booking_status']

# Create pairplot
print("Creating pairplot... This may take a moment.")
pairplot = sns.pairplot(df[key_features], hue='booking_status', 
                        palette={'Canceled': 'red', 'Not_Canceled': 'green'},
                        diag_kind='kde', plot_kws={'alpha': 0.6})
pairplot.fig.suptitle('Pairplot of Key Features', y=1.02, fontsize=16, fontweight='bold')
plt.show()
```
